## 🎯 Quick Navigation Helper

````python
def show_technique_info(technique_name: str):
    """
    Display information about a specific technique
    """
    technique_dir = Path('notebooks') / technique_name
    notebook_path = technique_dir / f"{technique_name}.ipynb"
    readme_path = technique_dir / "README.md"
    
    print(f"\n{'='*60}")
    print(f"📘 {technique_name.replace('_', ' ').title()}")
    print(f"{'='*60}\n")
    
    if notebook_path.exists():
        print(f"✅ Notebook: {notebook_path}")
    else:
        print(f"❌ Notebook not found: {notebook_path}")
    
    if readme_path.exists():
        print(f"✅ README: {readme_path}")
        print("\nTo view in Jupyter, open the README.md file or run:")
        print(f"   !cat {readme_path}")
    
    print(f"\n🚀 To open the notebook, navigate to:")
    print(f"   {notebook_path}")
    print()

# Example usage:
# show_technique_info('simple_rag')
# show_technique_info('graph_rag')
# show_technique_info('Agentic_RAG')
````
